```python
import matplotlib.pyplot as plt
import pandas as pd
import sqlalchemy
import numpy as np

%matplotlib inline
# engine = sqlalchemy.create_engine('mysql+pymysql://jcz:yub.miha@localhost:3306/weather')
engine = sqlalchemy.create_engine('mysql+pymysql://jcz:yub.miha@localhost:3306/weather')
boo = pd.read_sql_table("v_KPWM_daily",engine)
print(boo.dtypes)
boo.tail()
```

```python
df1 = pd.read_sql_table("v_monthly_gas",engine)
print(df1.dtypes)
df1.tail()
print("----------")

plt.figure(figsize=(30,10))
width = 0.35

plt.bar('YEARMONTH','gas_usage',width, color="red", data=df1)
#plt.bar('YEARMONTH','m_kwh', width, color="blue", data=df0)

plt.bar(np.arange(0,len(df1)),'gas_usage', width, color="red", data=df1)

# plt.xticks(np.arange(0,len(df1))+.25)
plt.xlabel("Year and Month")
plt.ylabel("Gas usage (cu ft) and Electric Usage (kwh)")
plt.title("Monthly Heat Utility Usage")
plt.xticks(rotation='vertical')

# Pad margins so that markers don't get clipped by the axes
plt.margins(0.01)
# Tweak spacing to prevent clipping of tick-labels
plt.savefig('HeatUtilsGasOnly.png')
plt.show()
```
